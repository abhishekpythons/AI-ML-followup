```python
import math
import random


def computePI (numThrows):
  in_circle = 0
  throws = 0

  while (throws < numThrows):
    xPos = random.uniform (-1.0, 1.0)
    yPos = random.uniform (-1.0, 1.0)
    if math.hypot (xPos, yPos) <= 1:
      in_circle += 1
    throws += 1
  pi = 4*in_circle/numThrows
  return pi


def main ():

  throws = (100, 1000, 10000, 100000, 1000000, 10000000)
  for numThrows in throws[0:7]:
    print(computePI(numThrows))
main ()
```

```python
from matplotlib import pyplot as plt
import random
import numpy as np
```

```python
n=30000
r=1000

incircle = 0

fig = plt.figure()
ax = fig.add_subplot()
ax.set_aspect('equal', adjustable='box')
for i in range(n):
  x = random.randint(-r,r)
  y = random.randint(-r,r)
  dist = (x**2+y**2)**0.5
  if dist>r:
    plt.plot(x, y, 'bo')
  else:
    plt.plot(x, y, 'ro')
    incircle += 1

plt.show()
print(4 * incircle/n)
```
